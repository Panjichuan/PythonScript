```python
import xlrd
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
sheet=xlrd.open_workbook(r'C:\Users\Administrator\OneDrive\Python\2017_物探项目提成分配情况_.xlsx')
table=sheet.sheet_by_name(sheet.sheet_names()[0])
data=[]
for i in np.arange(table.nrows):
    data.append(table.row_values(i))
df=DataFrame(data[1:],columns=data[0])
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置默认字体格式
ax1=df['分配金额/元'].plot(kind='bar')
ax1.set_xticklabels(df['姓名'],rotation=0)
plt.ylabel(u'金额')
plt.legend()
plt.show()
ax2=plt.pie(df['分配金额/元'],labels=df['姓名'],explode=(0.06,0.05,0.05,0.04),labeldistance=1.1,shadow=True,startangle=90)
plt.axis('equal')
plt.grid()
plt.legend(loc='best',bbox_to_anchor=(0.2,1))
plt.show()
```
