```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
A1=np.array([0,0])
B1=np.array(([2,0],[0,2]))
C1=np.random.multivariate_normal(A1,B1,10)
C2=np.random.multivariate_normal(A1+0.2,B1+0.2,10)
```

```python
fig1=plt.figure(figsize=(8,6))
plt.scatter(C1[:,0],C1[:,1],s=30,c='r',marker='o',alpha=0.5,label='C1')
plt.scatter(C2[:,0],C2[:,1],s=30,c='b',marker='x',alpha=0.5,label='C2')
plt.title('basic scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```

```python
x_coords=[0.13,0.22,0.39,0.59,0.68,0.74,0.93]
y_coords=[0.75,0.34,0.44,0.52,0.80,0.25,0.55]
```

```python
fig2=plt.figure(figsize=(8,5))
plt.scatter(x_coords,y_coords,marker='s',s=50)
for x,y in zip(x_coords,y_coords):
    plt.annotate('(%s,%s)'%(x,y),xy=(x,y),xytext=(0,-10),textcoords='offset points',ha='center',va='top')
plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```

```python
mu_vec1=np.array([0,0])
cov_mat1=np.array([[1,0],[0,1]])
X=np.random.multivariate_normal(mu_vec1,cov_mat1,500)
R=X**2
R_sum=R.sum(axis=1)
```

```python
fig3=plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],color='r',marker='o',s=32.*R_sum,edgecolors='r',alpha=0.5)
plt.show()
```
